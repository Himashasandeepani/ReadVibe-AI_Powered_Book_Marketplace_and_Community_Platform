```python
import csv
import pandas as pd

# 1) Read everything safely using csv.DictReader
records = []
with open("/content/amazon_books_reviews_merged_2014_2025.csv", "r", encoding="utf-8", errors="replace") as f:
    reader = csv.DictReader(f)
    for row in reader:
        records.append(row)

# Convert to DataFrame
df = pd.DataFrame(records)

# 2) Keep only User_id, profileName, and book Id
df_small = df[["User_id", "profileName", "Id"]].copy()

# Clean whitespace
df_small["User_id"]     = df_small["User_id"].astype(str).str.strip()
df_small["profileName"] = df_small["profileName"].astype(str).str.strip()
df_small["Id"]          = df_small["Id"].astype(str).str.strip()

# 3) Group by user and count unique books
user_book_counts = (
    df_small.groupby(["User_id", "profileName"])["Id"]
    .nunique()  # count unique book Ids
    .reset_index()
    .rename(columns={"Id": "unique_book_count"})
)

# 4) Filter users with more than one unique book
multi_book_users = user_book_counts[user_book_counts["unique_book_count"] > 1]

# 5) Print result
print("Users who have more than one book:")
print(multi_book_users)

# Optional: Show top users by count
print("\nTop users by number of distinct books:")
print(multi_book_users.sort_values(by="unique_book_count", ascending=False).head(20))
```

```python
import csv
import pandas as pd

# 1) Read everything safely using csv.DictReader
records = []
with open("/content/amazon_books_reviews_merged_2014_2025.csv", "r", encoding="utf-8", errors="replace") as f:
    reader = csv.DictReader(f)
    for row in reader:
        records.append(row)

# Convert to DataFrame
df = pd.DataFrame(records)

# 2) Keep only the relevant columns
df_filtered = df[["User_id", "profileName", "Id"]].copy()

# Clean whitespace
df_filtered["User_id"]     = df_filtered["User_id"].astype(str).str.strip()
df_filtered["profileName"] = df_filtered["profileName"].astype(str).str.strip()
df_filtered["Id"]          = df_filtered["Id"].astype(str).str.strip()

# 3) Group by user and collect unique book IDs
user_books = (
    df_filtered.groupby(["User_id", "profileName"])["Id"]
    .agg(lambda books: list(set(books)))  # list of distinct book IDs
    .reset_index()
    .rename(columns={"Id": "book_ids"})
)

# 4) Keep only users with more than 1 book
multi_book_users = user_books[user_books["book_ids"].map(len) > 1].copy()

# 5) Print final result
print("Users with more than one distinct book and their book IDs:")
print(multi_book_users)

# 6) Show how many user×book records we have
record_count = multi_book_users["book_ids"].map(len).sum()
print("\nTotal number of (user, book) associations after filtering:", record_count)
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd

# 1) List of transactions (users → books)
transactions = multi_book_users["book_ids"].tolist()

# 2) One-hot encode
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_binary = pd.DataFrame(te_ary, columns=te.columns_)

# 3) Frequent itemsets
freq_itemsets = apriori(df_binary, min_support=0.01, use_colnames=True)

print("Frequent itemsets:")
print(freq_itemsets.sort_values(by="support", ascending=False).head())

# 4) Extract rules
rules = association_rules(freq_itemsets, metric="confidence", min_threshold=0.5)

print("Association Rules:")
print(rules.sort_values(by="lift", ascending=False).head())
```

```python
# ===============================
# 1️⃣ IMPORT LIBRARIES
# ===============================

import csv
import random
import pandas as pd
from sklearn.model_selection import train_test_split
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules


# ===============================
# 2️⃣ LOAD DATA SAFELY
# ===============================

records = []
with open("/content/amazon_books_reviews_merged_2014_2025.csv",
          "r",
          encoding="utf-8",
          errors="replace") as f:
    reader = csv.DictReader(f)
    for row in reader:
        records.append(row)

df = pd.DataFrame(records)

print("Total raw rows loaded:", len(df))


# ===============================
# 3️⃣ PREPARE USER → BOOK LISTS
# ===============================

df_small = df[["User_id", "Id"]].copy()

df_small["User_id"] = df_small["User_id"].astype(str).str.strip()
df_small["Id"] = df_small["Id"].astype(str).str.strip()

# Remove missing
df_small = df_small.dropna()

# Group books per user
user_books = (
    df_small.groupby("User_id")["Id"]
    .agg(lambda x: list(set(x)))  # unique books per user
    .reset_index()
    .rename(columns={"Id": "book_ids"})
)

# Keep users with >1 book
multi_book_users = user_books[user_books["book_ids"].map(len) > 1].copy()

print("Users with >1 book:", len(multi_book_users))


# ===============================
# 4️⃣ CREATE TRANSACTIONS
# ===============================

transactions = multi_book_users["book_ids"].tolist()

print("Total transactions:", len(transactions))


# ===============================
# 5️⃣ TRAIN / TEST SPLIT (3:1)
# ===============================

train_transactions, test_transactions = train_test_split(
    transactions,
    test_size=0.25,
    random_state=42
)

print("Train users:", len(train_transactions))
print("Test users:", len(test_transactions))


# ===============================
# 6️⃣ TRAIN ASSOCIATION RULE MODEL
# ===============================

# One-hot encode
te = TransactionEncoder()
te_array = te.fit(train_transactions).transform(train_transactions)
train_df = pd.DataFrame(te_array, columns=te.columns_)

# Generate frequent itemsets
freq_itemsets = apriori(
    train_df,
    min_support=0.02,      # adjust if too few rules
    use_colnames=True
)

# Generate rules
rules = association_rules(
    freq_itemsets,
    metric="confidence",
    min_threshold=0.5
)

print("Frequent itemsets:", len(freq_itemsets))
print("Total rules generated:", len(rules))


# ===============================
# 7️⃣ RECOMMENDATION FUNCTION
# ===============================

def recommend_books(user_books, rules, top_k=5):

    recommendations = {}
    user_books = set(user_books)

    for _, row in rules.iterrows():

        antecedent = set(row["antecedents"])
        consequent = set(row["consequents"])

        if antecedent.issubset(user_books):

            for book in consequent:
                if book not in user_books:
                    recommendations[book] = recommendations.get(book, 0) + row["confidence"]

    # Sort by score
    sorted_books = sorted(recommendations.items(),
                          key=lambda x: x[1],
                          reverse=True)

    return [book for book, score in sorted_books[:top_k]]


# ===============================
# 8️⃣ EVALUATE MODEL (Hit Rate@K)
# ===============================

hits = 0
total = 0
K = 5

for user_books in test_transactions:

    if len(user_books) < 2:
        continue

    total += 1

    hidden_book = random.choice(user_books)
    input_books = list(set(user_books) - {hidden_book})

    recommended = recommend_books(input_books, rules, top_k=K)

    if hidden_book in recommended:
        hits += 1

hit_rate = hits / total if total > 0 else 0

print("\n===== EVALUATION RESULT =====")
print(f"Hit Rate@{K}: {hit_rate:.4f}")
print("Total evaluated users:", total)
print("Total hits:", hits)
```
